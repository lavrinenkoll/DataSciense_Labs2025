```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Зчитування даних з Excel, їх перегляд (перших 10 рядків)

```python
data = pd.read_excel('Data_Set_10.xls')
data
```

Видалення непотрібних стовпців: Row ID (порядок рядка), Country (усюди одна країна).\
Перетворення стовпців Order Date та Ship Date у формат дати.\
Сортування даних за датою замовлення.

```python
data = data.drop(columns=['Row ID', 'Country'])

data['Order Date'] = pd.to_datetime(data['Order Date'])
data['Ship Date'] = pd.to_datetime(data['Ship Date'])

data = data.sort_values(by='Order Date')

data
```

Обрані процеси для аналізу даних:\
Найприбутковіші та найзбитковіші підкатегорії товарів\
Найприбутковіші та найзбитковіші регіони\
Аналіз частоти покупок за сегментами клієнтів (Home Office, Corporate, Consumer)\
Найпопулярніші товари та вплив знижок на їхню купівельну активність

```python
def plot_bar_chart(x, y, title_x, title_y, title, color='skyblue'):
    plt.figure(figsize=(12, 6))
    plt.bar(x, y, color=color)
    plt.title(title)
    plt.xlabel(title_x)
    plt.ylabel(title_y)
    plt.xticks(rotation=45, ha='right')
    plt.grid(axis='y')
    plt.tight_layout()
    plt.show()

def statistical(y, title):
    mean = np.mean(y)
    median = np.median(y)
    std = np.std(y)
    min_val = np.min(y)
    max_val = np.max(y)

    print(f"\n📊 Статистичні характеристики: {title}")
    print(f"- Середнє значення: {mean:.2f}")
    print(f"- Медіана: {median:.2f}")
    print(f"- Стандартне відхилення: {std:.2f}")
    print(f"- Мінімальне значення: {min_val:.2f}")
    print(f"- Максимальне значення: {max_val:.2f}")

    # Висновки
    print("\n📝 Короткий висновок:")

    if std < 0.1 * mean:
        print("- Дані є стабільними (низьке коливання значень).")
    elif std > 0.3 * mean:
        print("- Дані мають значне коливання (висока нестабільність).")
    else:
        print("- Дані помірно варіативні.")

    if abs(mean - median) < 0.05 * mean:
        print("- Розподіл значень близький до симетричного.")
    elif mean > median:
        print("- Є зміщення розподілу вправо (в сторону більших значень).")
    else:
        print("- Є зміщення розподілу вліво (в сторону менших значень).")
```

```python
# найприбутковіші/найзбитковіші під-категорії товарів
category_profit = data.groupby('Sub-Category')['Profit'].sum().sort_values(ascending=False)
category_profit = category_profit.reset_index()
category_profit.columns = ['Sub-Category', 'Total Profit']
category_profit['Total Profit'] = category_profit['Total Profit'].round(2)

# Найприбутковіші під-категорії (тільки профіт > 0)
category_profit_positive = data[data['Profit'] > 0].groupby('Sub-Category')['Profit'].sum().sort_values(ascending=False)
category_profit_positive = category_profit_positive.reset_index()
category_profit_positive.columns = ['Sub-Category', 'Total Profit']
category_profit_positive['Total Profit'] = category_profit_positive['Total Profit'].round(2)
# Найзбитковіші під-категорії (тільки профіт < 0)
category_profit_negative = data[data['Profit'] < 0].groupby('Sub-Category')['Profit'].sum().sort_values(ascending=True)
category_profit_negative = category_profit_negative.reset_index()
category_profit_negative.columns = ['Sub-Category', 'Total Profit']
category_profit_negative['Total Profit'] = category_profit_negative['Total Profit'].round(2)

plot_bar_chart(category_profit['Sub-Category'], category_profit['Total Profit'],
               'Підкатегорія', 'Сумарний прибуток', 'Сумарний прибуток по підкатегоріях товарів')


# Об'єднуємо прибуткові та збиткові в один датафрейм для узгодженого сортування
profit_combined = pd.concat([category_profit_positive, category_profit_negative])
# Задаємо порядок підкатегорій як у першому графіку
profit_combined['Sub-Category'] = pd.Categorical(profit_combined['Sub-Category'], categories=category_profit['Sub-Category'], ordered=True)
profit_combined = profit_combined.sort_values('Sub-Category')

colors = profit_combined['Total Profit'].apply(lambda x: 'green' if x > 0 else 'red')
plot_bar_chart(profit_combined['Sub-Category'], profit_combined['Total Profit'],
               'Підкатегорія', 'Сумарний прибуток', 'Сумарний прибуток по підкатегоріях товарів (прибуток і збиток)',
               color=colors)
```

```python
# найприбутковіші/найзбитковіші регіони
region_profit = data.groupby('Region')['Profit'].sum().sort_values(ascending=False)
region_profit = region_profit.reset_index()
region_profit.columns = ['Region', 'Total Profit']
region_profit['Total Profit'] = region_profit['Total Profit'].round(2)

plot_bar_chart(region_profit['Region'], region_profit['Total Profit'],
               'Регіон', 'Сумарний прибуток', 'Сумарний прибуток по регіонах')


# Найприбутковіші регіони (тільки профіт > 0)
region_profit_positive = data[data['Profit'] > 0].groupby('Region')['Profit'].sum().sort_values(ascending=False)
region_profit_positive = region_profit_positive.reset_index()
region_profit_positive.columns = ['Region', 'Total Profit']
region_profit_positive['Total Profit'] = region_profit_positive['Total Profit'].round(2)
# Найзбитковіші регіони (тільки профіт < 0)
region_profit_negative = data[data['Profit'] < 0].groupby('Region')['Profit'].sum().sort_values(ascending=True)
region_profit_negative = region_profit_negative.reset_index()
region_profit_negative.columns = ['Region', 'Total Profit']
region_profit_negative['Total Profit'] = region_profit_negative['Total Profit'].round(2)
# Об'єднуємо прибуткові та збиткові в один датафрейм для узгодженого сортування
profit_combined = pd.concat([region_profit_positive, region_profit_negative])
# Задаємо порядок регіонів як у першому графіку
profit_combined['Region'] = pd.Categorical(profit_combined['Region'], categories=region_profit['Region'], ordered=True)
profit_combined = profit_combined.sort_values('Region')

# Побудова графіку
colors = profit_combined['Total Profit'].apply(lambda x: 'green' if x > 0 else 'red')
plot_bar_chart(profit_combined['Region'], profit_combined['Total Profit'],
               'Регіон', 'Сумарний прибуток', 'Сумарний прибуток по регіонах (прибуток і збиток)',
               color=colors)
```

```python
# часті покупці і звідки вони (segment: Home Office, Corporate, Consumer)
segment_count = data['Segment'].value_counts()
segment_count = segment_count.reset_index()
segment_count.columns = ['Segment', 'Count']

plot_bar_chart(segment_count['Segment'], segment_count['Count'],
               'Сегмент', 'Кількість покупців', 'Кількість покупців за сегментами')


# прибуткові/збиткові сегменти
segment_profit = data.groupby('Segment')['Profit'].sum().sort_values(ascending=False)
segment_profit = segment_profit.reset_index()
segment_profit.columns = ['Segment', 'Total Profit']
segment_profit['Total Profit'] = segment_profit['Total Profit'].round(2)

plot_bar_chart(segment_profit['Segment'], segment_profit['Total Profit'],
               'Сегмент', 'Сумарний прибуток', 'Сумарний прибуток по сегментах')


# Обчислення кореляції
correlation = segment_count['Count'].corr(segment_profit['Total Profit'])
print(f'Кореляція між кількістю покупців та сумарним прибутком за сегментами: {correlation:.2f}')

# Інтерпретація
if correlation > 0.7:
    print('Кореляція має високий показник, є сильна залежність: чим більше покупців у сегменті, тим вищий прибуток.')
elif correlation > 0.3:
    print('Кореляція середня: певна залежність присутня, але вона не абсолютна.')
elif correlation > 0:
    print('Кореляція слабка: прибуток і кількість покупців слабо пов’язані.')
else:
    print('Кореляція відсутня або негативна: залежності немає або вона зворотна.')

# Побудова графіка
merged_segment = segment_count.merge(segment_profit, on='Segment')
plt.figure(figsize=(8, 6))
sns.regplot(x='Count', y='Total Profit', data=merged_segment, marker='o', color='skyblue', line_kws={'color': 'red'})
plt.title('Залежність прибутку від кількості покупців за сегментами')
plt.xlabel('Кількість покупців')
plt.ylabel('Сумарний прибуток')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Найпопулярніші товари
product_count = data['Product Name'].value_counts()
product_count = product_count.reset_index()
product_count.columns = ['Product Name', 'Count']
product_count = product_count.sort_values(by='Count', ascending=False).head(30)

# Витягуємо повні назви популярних товарів
popular_names = product_count['Product Name']

# Обчислення відсотка знижок для популярних товарів
popular_products = data[data['Product Name'].isin(popular_names)]
discount_count = popular_products[popular_products['Discount'] > 0].groupby('Product Name')['Discount'].count()
discount_count = discount_count.reset_index()
discount_count.columns = ['Product Name', 'Discount Count']

# Об'єднуємо з product_count для збереження порядку
merged = product_count.merge(discount_count, on='Product Name', how='left')
merged['Discount Count'] = merged['Discount Count'].fillna(0)
merged['Discount Percentage'] = (merged['Discount Count'] / merged['Count']) * 100

# Створюємо окрему колонку зі скороченими назвами для візуалізації
merged['Short Name'] = merged['Product Name'].apply(lambda s: s[:15] + '...' if len(s) > 15 else s)

# Побудова графіків
plot_bar_chart(merged['Short Name'], merged['Count'],
               'Товар', 'Кількість покупок', 'Найчастіше куповані товари')

plot_bar_chart(merged['Short Name'], merged['Discount Percentage'],
               'Товар', 'Відсоток знижок > 0 (%)', 'Відсоток знижок > 0 для популярних товарів')


correlation = merged['Count'].corr(merged['Discount Percentage'])
print(f"Коефіцієнт кореляції між кількістю покупок і відсотком знижок: {correlation:.2f}")

# Інтерпретація
if correlation > 0.7:
    print("Сильна пряма залежність: чим частіше товар купують, тим частіше на нього дають знижки.")
elif correlation > 0.3:
    print("Помірна залежність: популярніші товари частіше мають знижки, але це не завжди так.")
elif correlation > 0:
    print("Слабка залежність: є певний зв'язок, але він незначний.")
else:
    print("Залежності майже немає або вона зворотна: частіше знижки дають на менш популярні товари.")
```

Побудова графіків змін.

```python
import warnings
from numpy import RankWarning
import matplotlib.dates as mdates


def find_best_polyfit(x, y, max_degree=10):
    best_degree = 1
    best_r2 = -np.inf
    with warnings.catch_warnings():
        warnings.simplefilter('ignore', RankWarning)
        for degree in range(1, max_degree + 1):
            coeffs = np.polyfit(x, y, degree)
            poly = np.poly1d(coeffs)
            y_pred = poly(x)
            ss_res = np.sum((y - y_pred) ** 2)
            ss_tot = np.sum((y - np.mean(y)) ** 2)
            r2 = 1 - ss_res / ss_tot
            if r2 > best_r2:
                best_r2 = r2
                best_degree = degree
    return best_degree, best_r2

def plot_time_series(data, title, ylabel, locator='month', max_degree=10):
    x = mdates.date2num(data.index)
    y = data.values

    trend_degree, r2 = find_best_polyfit(x, y, max_degree=max_degree)

    plt.figure(figsize=(12, 6))
    plt.plot(data.index, y, linestyle='-', label='Дані')
    plt.title(title)
    plt.xlabel('Дата')
    plt.ylabel(ylabel)
    plt.grid()

    ax = plt.gca()
    if locator == 'month':
        ax.xaxis.set_major_locator(mdates.MonthLocator())
        ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
    elif locator == 'week':
        ax.xaxis.set_major_locator(mdates.WeekdayLocator(interval=2))
        ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
    else:
        ax.xaxis.set_major_locator(mdates.DayLocator(interval=locator))
        ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))

    with warnings.catch_warnings():
        warnings.simplefilter('ignore', RankWarning)
        coeffs = np.polyfit(x, y, trend_degree)
    poly = np.poly1d(coeffs)
    trend = poly(x)

    plt.plot(data.index, trend, color='red', linewidth=2,
             label=f'Тренд (ступінь {trend_degree}, R²={r2:.3f})')

    plt.setp(ax.get_xticklabels(), rotation=90, ha='center', va='top')
    plt.legend()
    plt.tight_layout()
    plt.show()
```

```python
# Графік обсягу продажів
plot_time_series(sales_per_day, 'Обсяг продажів за день', 'Обсяг продажів')
```

```python
# видалення аномальних даних і перепобудова графіка
sales_per_day_no_anomaly = sales_per_day[sales_per_day < 10000]
plot_time_series(sales_per_day_no_anomaly, 'Обсяг продажів за день (без аномалій)', 'Обсяг продажів')
```

```python
# МНК прогнозування на N днів вперед + тренд на поточних даних
N_DAYS = 365
x = np.arange(len(sales_per_day_no_anomaly))
y = sales_per_day_no_anomaly.values

# Побудова полінома для тренду
coeffs = np.polyfit(x, y, 2)
poly = np.poly1d(coeffs)

# Прогноз на майбутнє
x_future = np.arange(len(sales_per_day_no_anomaly), len(sales_per_day_no_anomaly) + N_DAYS)
y_future = poly(x_future)
future_dates = pd.date_range(start=sales_per_day_no_anomaly.index[-1] + pd.Timedelta(days=1), periods=N_DAYS)

# Лінія тренду на поточних даних
y_trend = poly(x)

# Побудова графіка
plt.figure(figsize=(12, 6))
plt.plot(sales_per_day_no_anomaly.index, y, label='Дані')
plt.plot(sales_per_day_no_anomaly.index, y_trend, color='orange', linestyle='--', label='Тренд (МНК, 2)')
plt.plot(future_dates, y_future, color='red', label='Прогноз')
plt.title(f'Прогноз обсягу продажів на {N_DAYS} днів вперед')
plt.xlabel('Дата')
plt.ylabel('Обсяг продажів')
plt.grid()
plt.legend()
plt.gca().xaxis.set_major_locator(mdates.MonthLocator())
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
plt.setp(plt.gca().get_xticklabels(), rotation=90, ha='center', va='top')
plt.tight_layout()
plt.show()
```

```python
# Перевірка кореляції між різними метриками: обсяг продажів та кількість замовлень НА МІСЯЦЬ
sales_by_month = data.groupby('Month')['Sales'].sum()
orders_by_month = data.groupby('Month')['Order ID'].count()
correlation_orders = sales_by_month.corr(orders_by_month)
print(f'Кореляція між обсягом продажів та кількістю замовлень на місяць: {correlation_orders:.2f}')

# Перевірка кореляції між різними метриками: обсяг продажів та середній прибуток НА МІСЯЦЬ
average_profit_by_month = data.groupby('Month')['Profit'].mean()
correlation_profit = sales_by_month.corr(average_profit_by_month)
print(f'Кореляція між обсягом продажів та середнім прибутком на місяць: {correlation_profit:.2f}')

# Висновки
print("\nВисновки:")
if correlation_orders > 0.7:
    print("- Спостерігається сильна позитивна кореляція між кількістю замовлень і обсягом продажів. Це означає, що зі зростанням кількості замовлень зростає і обсяг продажів.")
elif correlation_orders > 0.3:
    print("- Існує помірна позитивна кореляція між кількістю замовлень і обсягом продажів.")
else:
    print("- Слабка або відсутня кореляція між кількістю замовлень і обсягом продажів — можливий вплив великих замовлень або сезонних факторів.")

if correlation_profit > 0.7:
    print("- Сильна позитивна кореляція між обсягом продажів і середнім прибутком — продажі супроводжуються високим прибутком.")
elif correlation_profit > 0.3:
    print("- Помірна позитивна кореляція між обсягом продажів і середнім прибутком — можливо, деякі товари продаються активно, але мають різний рівень прибутковості.")
else:
    print("- Слабка або відсутня кореляція між обсягом продажів і середнім прибутком — висока виручка не завжди означає високий прибуток.")
```
