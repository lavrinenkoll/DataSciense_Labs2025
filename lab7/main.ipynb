```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

–ó—á–∏—Ç—É–≤–∞–Ω–Ω—è –¥–∞–Ω–∏—Ö –∑ Excel, —ó—Ö –ø–µ—Ä–µ–≥–ª—è–¥ (–ø–µ—Ä—à–∏—Ö 10 —Ä—è–¥–∫—ñ–≤)

```python
data = pd.read_excel('Data_Set_10.xls')
data
```

–í–∏–¥–∞–ª–µ–Ω–Ω—è –Ω–µ–ø–æ—Ç—Ä—ñ–±–Ω–∏—Ö —Å—Ç–æ–≤–ø—Ü—ñ–≤: Row ID (–ø–æ—Ä—è–¥–æ–∫ —Ä—è–¥–∫–∞), Country (—É—Å—é–¥–∏ –æ–¥–Ω–∞ –∫—Ä–∞—ó–Ω–∞).\
–ü–µ—Ä–µ—Ç–≤–æ—Ä–µ–Ω–Ω—è —Å—Ç–æ–≤–ø—Ü—ñ–≤ Order Date —Ç–∞ Ship Date —É —Ñ–æ—Ä–º–∞—Ç –¥–∞—Ç–∏.\
–°–æ—Ä—Ç—É–≤–∞–Ω–Ω—è –¥–∞–Ω–∏—Ö –∑–∞ –¥–∞—Ç–æ—é –∑–∞–º–æ–≤–ª–µ–Ω–Ω—è.

```python
data = data.drop(columns=['Row ID', 'Country'])

data['Order Date'] = pd.to_datetime(data['Order Date'])
data['Ship Date'] = pd.to_datetime(data['Ship Date'])

data = data.sort_values(by='Order Date')

data
```

–û–±—Ä–∞–Ω—ñ –ø—Ä–æ—Ü–µ—Å–∏ –¥–ª—è –∞–Ω–∞–ª—ñ–∑—É –¥–∞–Ω–∏—Ö:\
–ù–∞–π–ø—Ä–∏–±—É—Ç–∫–æ–≤—ñ—à—ñ —Ç–∞ –Ω–∞–π–∑–±–∏—Ç–∫–æ–≤—ñ—à—ñ –ø—ñ–¥–∫–∞—Ç–µ–≥–æ—Ä—ñ—ó —Ç–æ–≤–∞—Ä—ñ–≤\
–ù–∞–π–ø—Ä–∏–±—É—Ç–∫–æ–≤—ñ—à—ñ —Ç–∞ –Ω–∞–π–∑–±–∏—Ç–∫–æ–≤—ñ—à—ñ —Ä–µ–≥—ñ–æ–Ω–∏\
–ê–Ω–∞–ª—ñ–∑ —á–∞—Å—Ç–æ—Ç–∏ –ø–æ–∫—É–ø–æ–∫ –∑–∞ —Å–µ–≥–º–µ–Ω—Ç–∞–º–∏ –∫–ª—ñ—î–Ω—Ç—ñ–≤ (Home Office, Corporate, Consumer)\
–ù–∞–π–ø–æ–ø—É–ª—è—Ä–Ω—ñ—à—ñ —Ç–æ–≤–∞—Ä–∏ —Ç–∞ –≤–ø–ª–∏–≤ –∑–Ω–∏–∂–æ–∫ –Ω–∞ —ó—Ö–Ω—é –∫—É–ø—ñ–≤–µ–ª—å–Ω—É –∞–∫—Ç–∏–≤–Ω—ñ—Å—Ç—å

```python
def plot_bar_chart(x, y, title_x, title_y, title, color='skyblue'):
    plt.figure(figsize=(12, 6))
    plt.bar(x, y, color=color)
    plt.title(title)
    plt.xlabel(title_x)
    plt.ylabel(title_y)
    plt.xticks(rotation=45, ha='right')
    plt.grid(axis='y')
    plt.tight_layout()
    plt.show()

def statistical(y, title):
    mean = np.mean(y)
    median = np.median(y)
    std = np.std(y)
    min_val = np.min(y)
    max_val = np.max(y)

    print(f"\nüìä –°—Ç–∞—Ç–∏—Å—Ç–∏—á–Ω—ñ —Ö–∞—Ä–∞–∫—Ç–µ—Ä–∏—Å—Ç–∏–∫–∏: {title}")
    print(f"- –°–µ—Ä–µ–¥–Ω—î –∑–Ω–∞—á–µ–Ω–Ω—è: {mean:.2f}")
    print(f"- –ú–µ–¥—ñ–∞–Ω–∞: {median:.2f}")
    print(f"- –°—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–µ –≤—ñ–¥—Ö–∏–ª–µ–Ω–Ω—è: {std:.2f}")
    print(f"- –ú—ñ–Ω—ñ–º–∞–ª—å–Ω–µ –∑–Ω–∞—á–µ–Ω–Ω—è: {min_val:.2f}")
    print(f"- –ú–∞–∫—Å–∏–º–∞–ª—å–Ω–µ –∑–Ω–∞—á–µ–Ω–Ω—è: {max_val:.2f}")

    # –í–∏—Å–Ω–æ–≤–∫–∏
    print("\nüìù –ö–æ—Ä–æ—Ç–∫–∏–π –≤–∏—Å–Ω–æ–≤–æ–∫:")

    if std < 0.1 * mean:
        print("- –î–∞–Ω—ñ —î —Å—Ç–∞–±—ñ–ª—å–Ω–∏–º–∏ (–Ω–∏–∑—å–∫–µ –∫–æ–ª–∏–≤–∞–Ω–Ω—è –∑–Ω–∞—á–µ–Ω—å).")
    elif std > 0.3 * mean:
        print("- –î–∞–Ω—ñ –º–∞—é—Ç—å –∑–Ω–∞—á–Ω–µ –∫–æ–ª–∏–≤–∞–Ω–Ω—è (–≤–∏—Å–æ–∫–∞ –Ω–µ—Å—Ç–∞–±—ñ–ª—å–Ω—ñ—Å—Ç—å).")
    else:
        print("- –î–∞–Ω—ñ –ø–æ–º—ñ—Ä–Ω–æ –≤–∞—Ä—ñ–∞—Ç–∏–≤–Ω—ñ.")

    if abs(mean - median) < 0.05 * mean:
        print("- –†–æ–∑–ø–æ–¥—ñ–ª –∑–Ω–∞—á–µ–Ω—å –±–ª–∏–∑—å–∫–∏–π –¥–æ —Å–∏–º–µ—Ç—Ä–∏—á–Ω–æ–≥–æ.")
    elif mean > median:
        print("- –Ñ –∑–º—ñ—â–µ–Ω–Ω—è —Ä–æ–∑–ø–æ–¥—ñ–ª—É –≤–ø—Ä–∞–≤–æ (–≤ —Å—Ç–æ—Ä–æ–Ω—É –±—ñ–ª—å—à–∏—Ö –∑–Ω–∞—á–µ–Ω—å).")
    else:
        print("- –Ñ –∑–º—ñ—â–µ–Ω–Ω—è —Ä–æ–∑–ø–æ–¥—ñ–ª—É –≤–ª—ñ–≤–æ (–≤ —Å—Ç–æ—Ä–æ–Ω—É –º–µ–Ω—à–∏—Ö –∑–Ω–∞—á–µ–Ω—å).")
```

```python
# –Ω–∞–π–ø—Ä–∏–±—É—Ç–∫–æ–≤—ñ—à—ñ/–Ω–∞–π–∑–±–∏—Ç–∫–æ–≤—ñ—à—ñ –ø—ñ–¥-–∫–∞—Ç–µ–≥–æ—Ä—ñ—ó —Ç–æ–≤–∞—Ä—ñ–≤
category_profit = data.groupby('Sub-Category')['Profit'].sum().sort_values(ascending=False)
category_profit = category_profit.reset_index()
category_profit.columns = ['Sub-Category', 'Total Profit']
category_profit['Total Profit'] = category_profit['Total Profit'].round(2)

# –ù–∞–π–ø—Ä–∏–±—É—Ç–∫–æ–≤—ñ—à—ñ –ø—ñ–¥-–∫–∞—Ç–µ–≥–æ—Ä—ñ—ó (—Ç—ñ–ª—å–∫–∏ –ø—Ä–æ—Ñ—ñ—Ç > 0)
category_profit_positive = data[data['Profit'] > 0].groupby('Sub-Category')['Profit'].sum().sort_values(ascending=False)
category_profit_positive = category_profit_positive.reset_index()
category_profit_positive.columns = ['Sub-Category', 'Total Profit']
category_profit_positive['Total Profit'] = category_profit_positive['Total Profit'].round(2)
# –ù–∞–π–∑–±–∏—Ç–∫–æ–≤—ñ—à—ñ –ø—ñ–¥-–∫–∞—Ç–µ–≥–æ—Ä—ñ—ó (—Ç—ñ–ª—å–∫–∏ –ø—Ä–æ—Ñ—ñ—Ç < 0)
category_profit_negative = data[data['Profit'] < 0].groupby('Sub-Category')['Profit'].sum().sort_values(ascending=True)
category_profit_negative = category_profit_negative.reset_index()
category_profit_negative.columns = ['Sub-Category', 'Total Profit']
category_profit_negative['Total Profit'] = category_profit_negative['Total Profit'].round(2)

plot_bar_chart(category_profit['Sub-Category'], category_profit['Total Profit'],
               '–ü—ñ–¥–∫–∞—Ç–µ–≥–æ—Ä—ñ—è', '–°—É–º–∞—Ä–Ω–∏–π –ø—Ä–∏–±—É—Ç–æ–∫', '–°—É–º–∞—Ä–Ω–∏–π –ø—Ä–∏–±—É—Ç–æ–∫ –ø–æ –ø—ñ–¥–∫–∞—Ç–µ–≥–æ—Ä—ñ—è—Ö —Ç–æ–≤–∞—Ä—ñ–≤')


# –û–±'—î–¥–Ω—É—î–º–æ –ø—Ä–∏–±—É—Ç–∫–æ–≤—ñ —Ç–∞ –∑–±–∏—Ç–∫–æ–≤—ñ –≤ –æ–¥–∏–Ω –¥–∞—Ç–∞—Ñ—Ä–µ–π–º –¥–ª—è —É–∑–≥–æ–¥–∂–µ–Ω–æ–≥–æ —Å–æ—Ä—Ç—É–≤–∞–Ω–Ω—è
profit_combined = pd.concat([category_profit_positive, category_profit_negative])
# –ó–∞–¥–∞—î–º–æ –ø–æ—Ä—è–¥–æ–∫ –ø—ñ–¥–∫–∞—Ç–µ–≥–æ—Ä—ñ–π —è–∫ —É –ø–µ—Ä—à–æ–º—É –≥—Ä–∞—Ñ—ñ–∫—É
profit_combined['Sub-Category'] = pd.Categorical(profit_combined['Sub-Category'], categories=category_profit['Sub-Category'], ordered=True)
profit_combined = profit_combined.sort_values('Sub-Category')

colors = profit_combined['Total Profit'].apply(lambda x: 'green' if x > 0 else 'red')
plot_bar_chart(profit_combined['Sub-Category'], profit_combined['Total Profit'],
               '–ü—ñ–¥–∫–∞—Ç–µ–≥–æ—Ä—ñ—è', '–°—É–º–∞—Ä–Ω–∏–π –ø—Ä–∏–±—É—Ç–æ–∫', '–°—É–º–∞—Ä–Ω–∏–π –ø—Ä–∏–±—É—Ç–æ–∫ –ø–æ –ø—ñ–¥–∫–∞—Ç–µ–≥–æ—Ä—ñ—è—Ö —Ç–æ–≤–∞—Ä—ñ–≤ (–ø—Ä–∏–±—É—Ç–æ–∫ —ñ –∑–±–∏—Ç–æ–∫)',
               color=colors)
```

```python
# –Ω–∞–π–ø—Ä–∏–±—É—Ç–∫–æ–≤—ñ—à—ñ/–Ω–∞–π–∑–±–∏—Ç–∫–æ–≤—ñ—à—ñ —Ä–µ–≥—ñ–æ–Ω–∏
region_profit = data.groupby('Region')['Profit'].sum().sort_values(ascending=False)
region_profit = region_profit.reset_index()
region_profit.columns = ['Region', 'Total Profit']
region_profit['Total Profit'] = region_profit['Total Profit'].round(2)

plot_bar_chart(region_profit['Region'], region_profit['Total Profit'],
               '–†–µ–≥—ñ–æ–Ω', '–°—É–º–∞—Ä–Ω–∏–π –ø—Ä–∏–±—É—Ç–æ–∫', '–°—É–º–∞—Ä–Ω–∏–π –ø—Ä–∏–±—É—Ç–æ–∫ –ø–æ —Ä–µ–≥—ñ–æ–Ω–∞—Ö')


# –ù–∞–π–ø—Ä–∏–±—É—Ç–∫–æ–≤—ñ—à—ñ —Ä–µ–≥—ñ–æ–Ω–∏ (—Ç—ñ–ª—å–∫–∏ –ø—Ä–æ—Ñ—ñ—Ç > 0)
region_profit_positive = data[data['Profit'] > 0].groupby('Region')['Profit'].sum().sort_values(ascending=False)
region_profit_positive = region_profit_positive.reset_index()
region_profit_positive.columns = ['Region', 'Total Profit']
region_profit_positive['Total Profit'] = region_profit_positive['Total Profit'].round(2)
# –ù–∞–π–∑–±–∏—Ç–∫–æ–≤—ñ—à—ñ —Ä–µ–≥—ñ–æ–Ω–∏ (—Ç—ñ–ª—å–∫–∏ –ø—Ä–æ—Ñ—ñ—Ç < 0)
region_profit_negative = data[data['Profit'] < 0].groupby('Region')['Profit'].sum().sort_values(ascending=True)
region_profit_negative = region_profit_negative.reset_index()
region_profit_negative.columns = ['Region', 'Total Profit']
region_profit_negative['Total Profit'] = region_profit_negative['Total Profit'].round(2)
# –û–±'—î–¥–Ω—É—î–º–æ –ø—Ä–∏–±—É—Ç–∫–æ–≤—ñ —Ç–∞ –∑–±–∏—Ç–∫–æ–≤—ñ –≤ –æ–¥–∏–Ω –¥–∞—Ç–∞—Ñ—Ä–µ–π–º –¥–ª—è —É–∑–≥–æ–¥–∂–µ–Ω–æ–≥–æ —Å–æ—Ä—Ç—É–≤–∞–Ω–Ω—è
profit_combined = pd.concat([region_profit_positive, region_profit_negative])
# –ó–∞–¥–∞—î–º–æ –ø–æ—Ä—è–¥–æ–∫ —Ä–µ–≥—ñ–æ–Ω—ñ–≤ —è–∫ —É –ø–µ—Ä—à–æ–º—É –≥—Ä–∞—Ñ—ñ–∫—É
profit_combined['Region'] = pd.Categorical(profit_combined['Region'], categories=region_profit['Region'], ordered=True)
profit_combined = profit_combined.sort_values('Region')

# –ü–æ–±—É–¥–æ–≤–∞ –≥—Ä–∞—Ñ—ñ–∫—É
colors = profit_combined['Total Profit'].apply(lambda x: 'green' if x > 0 else 'red')
plot_bar_chart(profit_combined['Region'], profit_combined['Total Profit'],
               '–†–µ–≥—ñ–æ–Ω', '–°—É–º–∞—Ä–Ω–∏–π –ø—Ä–∏–±—É—Ç–æ–∫', '–°—É–º–∞—Ä–Ω–∏–π –ø—Ä–∏–±—É—Ç–æ–∫ –ø–æ —Ä–µ–≥—ñ–æ–Ω–∞—Ö (–ø—Ä–∏–±—É—Ç–æ–∫ —ñ –∑–±–∏—Ç–æ–∫)',
               color=colors)
```

```python
# —á–∞—Å—Ç—ñ –ø–æ–∫—É–ø—Ü—ñ —ñ –∑–≤—ñ–¥–∫–∏ –≤–æ–Ω–∏ (segment: Home Office, Corporate, Consumer)
segment_count = data['Segment'].value_counts()
segment_count = segment_count.reset_index()
segment_count.columns = ['Segment', 'Count']

plot_bar_chart(segment_count['Segment'], segment_count['Count'],
               '–°–µ–≥–º–µ–Ω—Ç', '–ö—ñ–ª—å–∫—ñ—Å—Ç—å –ø–æ–∫—É–ø—Ü—ñ–≤', '–ö—ñ–ª—å–∫—ñ—Å—Ç—å –ø–æ–∫—É–ø—Ü—ñ–≤ –∑–∞ —Å–µ–≥–º–µ–Ω—Ç–∞–º–∏')


# –ø—Ä–∏–±—É—Ç–∫–æ–≤—ñ/–∑–±–∏—Ç–∫–æ–≤—ñ —Å–µ–≥–º–µ–Ω—Ç–∏
segment_profit = data.groupby('Segment')['Profit'].sum().sort_values(ascending=False)
segment_profit = segment_profit.reset_index()
segment_profit.columns = ['Segment', 'Total Profit']
segment_profit['Total Profit'] = segment_profit['Total Profit'].round(2)

plot_bar_chart(segment_profit['Segment'], segment_profit['Total Profit'],
               '–°–µ–≥–º–µ–Ω—Ç', '–°—É–º–∞—Ä–Ω–∏–π –ø—Ä–∏–±—É—Ç–æ–∫', '–°—É–º–∞—Ä–Ω–∏–π –ø—Ä–∏–±—É—Ç–æ–∫ –ø–æ —Å–µ–≥–º–µ–Ω—Ç–∞—Ö')


# –û–±—á–∏—Å–ª–µ–Ω–Ω—è –∫–æ—Ä–µ–ª—è—Ü—ñ—ó
correlation = segment_count['Count'].corr(segment_profit['Total Profit'])
print(f'–ö–æ—Ä–µ–ª—è—Ü—ñ—è –º—ñ–∂ –∫—ñ–ª—å–∫—ñ—Å—Ç—é –ø–æ–∫—É–ø—Ü—ñ–≤ —Ç–∞ —Å—É–º–∞—Ä–Ω–∏–º –ø—Ä–∏–±—É—Ç–∫–æ–º –∑–∞ —Å–µ–≥–º–µ–Ω—Ç–∞–º–∏: {correlation:.2f}')

# –Ü–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∞—Ü—ñ—è
if correlation > 0.7:
    print('–ö–æ—Ä–µ–ª—è—Ü—ñ—è –º–∞—î –≤–∏—Å–æ–∫–∏–π –ø–æ–∫–∞–∑–Ω–∏–∫, —î —Å–∏–ª—å–Ω–∞ –∑–∞–ª–µ–∂–Ω—ñ—Å—Ç—å: —á–∏–º –±—ñ–ª—å—à–µ –ø–æ–∫—É–ø—Ü—ñ–≤ —É —Å–µ–≥–º–µ–Ω—Ç—ñ, —Ç–∏–º –≤–∏—â–∏–π –ø—Ä–∏–±—É—Ç–æ–∫.')
elif correlation > 0.3:
    print('–ö–æ—Ä–µ–ª—è—Ü—ñ—è —Å–µ—Ä–µ–¥–Ω—è: –ø–µ–≤–Ω–∞ –∑–∞–ª–µ–∂–Ω—ñ—Å—Ç—å –ø—Ä–∏—Å—É—Ç–Ω—è, –∞–ª–µ –≤–æ–Ω–∞ –Ω–µ –∞–±—Å–æ–ª—é—Ç–Ω–∞.')
elif correlation > 0:
    print('–ö–æ—Ä–µ–ª—è—Ü—ñ—è —Å–ª–∞–±–∫–∞: –ø—Ä–∏–±—É—Ç–æ–∫ —ñ –∫—ñ–ª—å–∫—ñ—Å—Ç—å –ø–æ–∫—É–ø—Ü—ñ–≤ —Å–ª–∞–±–æ –ø–æ–≤‚Äô—è–∑–∞–Ω—ñ.')
else:
    print('–ö–æ—Ä–µ–ª—è—Ü—ñ—è –≤—ñ–¥—Å—É—Ç–Ω—è –∞–±–æ –Ω–µ–≥–∞—Ç–∏–≤–Ω–∞: –∑–∞–ª–µ–∂–Ω–æ—Å—Ç—ñ –Ω–µ–º–∞—î –∞–±–æ –≤–æ–Ω–∞ –∑–≤–æ—Ä–æ—Ç–Ω–∞.')

# –ü–æ–±—É–¥–æ–≤–∞ –≥—Ä–∞—Ñ—ñ–∫–∞
merged_segment = segment_count.merge(segment_profit, on='Segment')
plt.figure(figsize=(8, 6))
sns.regplot(x='Count', y='Total Profit', data=merged_segment, marker='o', color='skyblue', line_kws={'color': 'red'})
plt.title('–ó–∞–ª–µ–∂–Ω—ñ—Å—Ç—å –ø—Ä–∏–±—É—Ç–∫—É –≤—ñ–¥ –∫—ñ–ª—å–∫–æ—Å—Ç—ñ –ø–æ–∫—É–ø—Ü—ñ–≤ –∑–∞ —Å–µ–≥–º–µ–Ω—Ç–∞–º–∏')
plt.xlabel('–ö—ñ–ª—å–∫—ñ—Å—Ç—å –ø–æ–∫—É–ø—Ü—ñ–≤')
plt.ylabel('–°—É–º–∞—Ä–Ω–∏–π –ø—Ä–∏–±—É—Ç–æ–∫')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# –ù–∞–π–ø–æ–ø—É–ª—è—Ä–Ω—ñ—à—ñ —Ç–æ–≤–∞—Ä–∏
product_count = data['Product Name'].value_counts()
product_count = product_count.reset_index()
product_count.columns = ['Product Name', 'Count']
product_count = product_count.sort_values(by='Count', ascending=False).head(30)

# –í–∏—Ç—è–≥—É—î–º–æ –ø–æ–≤–Ω—ñ –Ω–∞–∑–≤–∏ –ø–æ–ø—É–ª—è—Ä–Ω–∏—Ö —Ç–æ–≤–∞—Ä—ñ–≤
popular_names = product_count['Product Name']

# –û–±—á–∏—Å–ª–µ–Ω–Ω—è –≤—ñ–¥—Å–æ—Ç–∫–∞ –∑–Ω–∏–∂–æ–∫ –¥–ª—è –ø–æ–ø—É–ª—è—Ä–Ω–∏—Ö —Ç–æ–≤–∞—Ä—ñ–≤
popular_products = data[data['Product Name'].isin(popular_names)]
discount_count = popular_products[popular_products['Discount'] > 0].groupby('Product Name')['Discount'].count()
discount_count = discount_count.reset_index()
discount_count.columns = ['Product Name', 'Discount Count']

# –û–±'—î–¥–Ω—É—î–º–æ –∑ product_count –¥–ª—è –∑–±–µ—Ä–µ–∂–µ–Ω–Ω—è –ø–æ—Ä—è–¥–∫—É
merged = product_count.merge(discount_count, on='Product Name', how='left')
merged['Discount Count'] = merged['Discount Count'].fillna(0)
merged['Discount Percentage'] = (merged['Discount Count'] / merged['Count']) * 100

# –°—Ç–≤–æ—Ä—é—î–º–æ –æ–∫—Ä–µ–º—É –∫–æ–ª–æ–Ω–∫—É –∑—ñ —Å–∫–æ—Ä–æ—á–µ–Ω–∏–º–∏ –Ω–∞–∑–≤–∞–º–∏ –¥–ª—è –≤—ñ–∑—É–∞–ª—ñ–∑–∞—Ü—ñ—ó
merged['Short Name'] = merged['Product Name'].apply(lambda s: s[:15] + '...' if len(s) > 15 else s)

# –ü–æ–±—É–¥–æ–≤–∞ –≥—Ä–∞—Ñ—ñ–∫—ñ–≤
plot_bar_chart(merged['Short Name'], merged['Count'],
               '–¢–æ–≤–∞—Ä', '–ö—ñ–ª—å–∫—ñ—Å—Ç—å –ø–æ–∫—É–ø–æ–∫', '–ù–∞–π—á–∞—Å—Ç—ñ—à–µ –∫—É–ø–æ–≤–∞–Ω—ñ —Ç–æ–≤–∞—Ä–∏')

plot_bar_chart(merged['Short Name'], merged['Discount Percentage'],
               '–¢–æ–≤–∞—Ä', '–í—ñ–¥—Å–æ—Ç–æ–∫ –∑–Ω–∏–∂–æ–∫ > 0 (%)', '–í—ñ–¥—Å–æ—Ç–æ–∫ –∑–Ω–∏–∂–æ–∫ > 0 –¥–ª—è –ø–æ–ø—É–ª—è—Ä–Ω–∏—Ö —Ç–æ–≤–∞—Ä—ñ–≤')


correlation = merged['Count'].corr(merged['Discount Percentage'])
print(f"–ö–æ–µ—Ñ—ñ—Ü—ñ—î–Ω—Ç –∫–æ—Ä–µ–ª—è—Ü—ñ—ó –º—ñ–∂ –∫—ñ–ª—å–∫—ñ—Å—Ç—é –ø–æ–∫—É–ø–æ–∫ —ñ –≤—ñ–¥—Å–æ—Ç–∫–æ–º –∑–Ω–∏–∂–æ–∫: {correlation:.2f}")

# –Ü–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∞—Ü—ñ—è
if correlation > 0.7:
    print("–°–∏–ª—å–Ω–∞ –ø—Ä—è–º–∞ –∑–∞–ª–µ–∂–Ω—ñ—Å—Ç—å: —á–∏–º —á–∞—Å—Ç—ñ—à–µ —Ç–æ–≤–∞—Ä –∫—É–ø—É—é—Ç—å, —Ç–∏–º —á–∞—Å—Ç—ñ—à–µ –Ω–∞ –Ω—å–æ–≥–æ –¥–∞—é—Ç—å –∑–Ω–∏–∂–∫–∏.")
elif correlation > 0.3:
    print("–ü–æ–º—ñ—Ä–Ω–∞ –∑–∞–ª–µ–∂–Ω—ñ—Å—Ç—å: –ø–æ–ø—É–ª—è—Ä–Ω—ñ—à—ñ —Ç–æ–≤–∞—Ä–∏ —á–∞—Å—Ç—ñ—à–µ –º–∞—é—Ç—å –∑–Ω–∏–∂–∫–∏, –∞–ª–µ —Ü–µ –Ω–µ –∑–∞–≤–∂–¥–∏ —Ç–∞–∫.")
elif correlation > 0:
    print("–°–ª–∞–±–∫–∞ –∑–∞–ª–µ–∂–Ω—ñ—Å—Ç—å: —î –ø–µ–≤–Ω–∏–π –∑–≤'—è–∑–æ–∫, –∞–ª–µ –≤—ñ–Ω –Ω–µ–∑–Ω–∞—á–Ω–∏–π.")
else:
    print("–ó–∞–ª–µ–∂–Ω–æ—Å—Ç—ñ –º–∞–π–∂–µ –Ω–µ–º–∞—î –∞–±–æ –≤–æ–Ω–∞ –∑–≤–æ—Ä–æ—Ç–Ω–∞: —á–∞—Å—Ç—ñ—à–µ –∑–Ω–∏–∂–∫–∏ –¥–∞—é—Ç—å –Ω–∞ –º–µ–Ω—à –ø–æ–ø—É–ª—è—Ä–Ω—ñ —Ç–æ–≤–∞—Ä–∏.")
```

```python
import warnings
from numpy import RankWarning
import matplotlib.dates as mdates


def find_best_polyfit(x, y, max_degree=10):
    best_degree = 1
    best_r2 = -np.inf
    with warnings.catch_warnings():
        warnings.simplefilter('ignore', RankWarning)
        for degree in range(1, max_degree + 1):
            coeffs = np.polyfit(x, y, degree)
            poly = np.poly1d(coeffs)
            y_pred = poly(x)
            ss_res = np.sum((y - y_pred) ** 2)
            ss_tot = np.sum((y - np.mean(y)) ** 2)
            r2 = 1 - ss_res / ss_tot
            if r2 > best_r2:
                best_r2 = r2
                best_degree = degree
    return best_degree, best_r2

def plot_time_series(data, title, ylabel, locator='month', max_degree=10):
    x = mdates.date2num(data.index)
    y = data.values

    trend_degree, r2 = find_best_polyfit(x, y, max_degree=max_degree)

    plt.figure(figsize=(12, 6))
    plt.plot(data.index, y, linestyle='-', label='–î–∞–Ω—ñ')
    plt.title(title)
    plt.xlabel('–î–∞—Ç–∞')
    plt.ylabel(ylabel)
    plt.grid()

    ax = plt.gca()
    if locator == 'month':
        ax.xaxis.set_major_locator(mdates.MonthLocator())
        ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
    elif locator == 'week':
        ax.xaxis.set_major_locator(mdates.WeekdayLocator(interval=2))
        ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
    else:
        ax.xaxis.set_major_locator(mdates.DayLocator(interval=locator))
        ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))

    with warnings.catch_warnings():
        warnings.simplefilter('ignore', RankWarning)
        coeffs = np.polyfit(x, y, trend_degree)
    poly = np.poly1d(coeffs)
    trend = poly(x)

    plt.plot(data.index, trend, color='red', linewidth=2,
             label=f'–¢—Ä–µ–Ω–¥ (—Å—Ç—É–ø—ñ–Ω—å {trend_degree}, R¬≤={r2:.3f})')

    plt.setp(ax.get_xticklabels(), rotation=90, ha='center', va='top')
    plt.legend()
    plt.tight_layout()
    plt.show()
```

```python
# –ü–µ—Ä–µ–≤—ñ—Ä–∫–∞ –∫–æ—Ä–µ–ª—è—Ü—ñ—ó –º—ñ–∂ —Ä—ñ–∑–Ω–∏–º–∏ –º–µ—Ç—Ä–∏–∫–∞–º–∏: –æ–±—Å—è–≥ –ø—Ä–æ–¥–∞–∂—ñ–≤ —Ç–∞ –∫—ñ–ª—å–∫—ñ—Å—Ç—å –∑–∞–º–æ–≤–ª–µ–Ω—å –ù–ê –ú–Ü–°–Ø–¶–¨
sales_by_month = data.groupby('Month')['Sales'].sum()
orders_by_month = data.groupby('Month')['Order ID'].count()
correlation_orders = sales_by_month.corr(orders_by_month)
print(f'–ö–æ—Ä–µ–ª—è—Ü—ñ—è –º—ñ–∂ –æ–±—Å—è–≥–æ–º –ø—Ä–æ–¥–∞–∂—ñ–≤ —Ç–∞ –∫—ñ–ª—å–∫—ñ—Å—Ç—é –∑–∞–º–æ–≤–ª–µ–Ω—å –Ω–∞ –º—ñ—Å—è—Ü—å: {correlation_orders:.2f}')

# –ü–µ—Ä–µ–≤—ñ—Ä–∫–∞ –∫–æ—Ä–µ–ª—è—Ü—ñ—ó –º—ñ–∂ —Ä—ñ–∑–Ω–∏–º–∏ –º–µ—Ç—Ä–∏–∫–∞–º–∏: –æ–±—Å—è–≥ –ø—Ä–æ–¥–∞–∂—ñ–≤ —Ç–∞ —Å–µ—Ä–µ–¥–Ω—ñ–π –ø—Ä–∏–±—É—Ç–æ–∫ –ù–ê –ú–Ü–°–Ø–¶–¨
average_profit_by_month = data.groupby('Month')['Profit'].mean()
correlation_profit = sales_by_month.corr(average_profit_by_month)
print(f'–ö–æ—Ä–µ–ª—è—Ü—ñ—è –º—ñ–∂ –æ–±—Å—è–≥–æ–º –ø—Ä–æ–¥–∞–∂—ñ–≤ —Ç–∞ —Å–µ—Ä–µ–¥–Ω—ñ–º –ø—Ä–∏–±—É—Ç–∫–æ–º –Ω–∞ –º—ñ—Å—è—Ü—å: {correlation_profit:.2f}')

# –í–∏—Å–Ω–æ–≤–∫–∏
print("\n–í–∏—Å–Ω–æ–≤–∫–∏:")
if correlation_orders > 0.7:
    print("- –°–ø–æ—Å—Ç–µ—Ä—ñ–≥–∞—î—Ç—å—Å—è —Å–∏–ª—å–Ω–∞ –ø–æ–∑–∏—Ç–∏–≤–Ω–∞ –∫–æ—Ä–µ–ª—è—Ü—ñ—è –º—ñ–∂ –∫—ñ–ª—å–∫—ñ—Å—Ç—é –∑–∞–º–æ–≤–ª–µ–Ω—å —ñ –æ–±—Å—è–≥–æ–º –ø—Ä–æ–¥–∞–∂—ñ–≤. –¶–µ –æ–∑–Ω–∞—á–∞—î, —â–æ –∑—ñ –∑—Ä–æ—Å—Ç–∞–Ω–Ω—è–º –∫—ñ–ª—å–∫–æ—Å—Ç—ñ –∑–∞–º–æ–≤–ª–µ–Ω—å –∑—Ä–æ—Å—Ç–∞—î —ñ –æ–±—Å—è–≥ –ø—Ä–æ–¥–∞–∂—ñ–≤.")
elif correlation_orders > 0.3:
    print("- –Ü—Å–Ω—É—î –ø–æ–º—ñ—Ä–Ω–∞ –ø–æ–∑–∏—Ç–∏–≤–Ω–∞ –∫–æ—Ä–µ–ª—è—Ü—ñ—è –º—ñ–∂ –∫—ñ–ª—å–∫—ñ—Å—Ç—é –∑–∞–º–æ–≤–ª–µ–Ω—å —ñ –æ–±—Å—è–≥–æ–º –ø—Ä–æ–¥–∞–∂—ñ–≤.")
else:
    print("- –°–ª–∞–±–∫–∞ –∞–±–æ –≤—ñ–¥—Å—É—Ç–Ω—è –∫–æ—Ä–µ–ª—è—Ü—ñ—è –º—ñ–∂ –∫—ñ–ª—å–∫—ñ—Å—Ç—é –∑–∞–º–æ–≤–ª–µ–Ω—å —ñ –æ–±—Å—è–≥–æ–º –ø—Ä–æ–¥–∞–∂—ñ–≤ ‚Äî –º–æ–∂–ª–∏–≤–∏–π –≤–ø–ª–∏–≤ –≤–µ–ª–∏–∫–∏—Ö –∑–∞–º–æ–≤–ª–µ–Ω—å –∞–±–æ —Å–µ–∑–æ–Ω–Ω–∏—Ö —Ñ–∞–∫—Ç–æ—Ä—ñ–≤.")

if correlation_profit > 0.7:
    print("- –°–∏–ª—å–Ω–∞ –ø–æ–∑–∏—Ç–∏–≤–Ω–∞ –∫–æ—Ä–µ–ª—è—Ü—ñ—è –º—ñ–∂ –æ–±—Å—è–≥–æ–º –ø—Ä–æ–¥–∞–∂—ñ–≤ —ñ —Å–µ—Ä–µ–¥–Ω—ñ–º –ø—Ä–∏–±—É—Ç–∫–æ–º ‚Äî –ø—Ä–æ–¥–∞–∂—ñ —Å—É–ø—Ä–æ–≤–æ–¥–∂—É—é—Ç—å—Å—è –≤–∏—Å–æ–∫–∏–º –ø—Ä–∏–±—É—Ç–∫–æ–º.")
elif correlation_profit > 0.3:
    print("- –ü–æ–º—ñ—Ä–Ω–∞ –ø–æ–∑–∏—Ç–∏–≤–Ω–∞ –∫–æ—Ä–µ–ª—è—Ü—ñ—è –º—ñ–∂ –æ–±—Å—è–≥–æ–º –ø—Ä–æ–¥–∞–∂—ñ–≤ —ñ —Å–µ—Ä–µ–¥–Ω—ñ–º –ø—Ä–∏–±—É—Ç–∫–æ–º ‚Äî –º–æ–∂–ª–∏–≤–æ, –¥–µ—è–∫—ñ —Ç–æ–≤–∞—Ä–∏ –ø—Ä–æ–¥–∞—é—Ç—å—Å—è –∞–∫—Ç–∏–≤–Ω–æ, –∞–ª–µ –º–∞—é—Ç—å —Ä—ñ–∑–Ω–∏–π —Ä—ñ–≤–µ–Ω—å –ø—Ä–∏–±—É—Ç–∫–æ–≤–æ—Å—Ç—ñ.")
else:
    print("- –°–ª–∞–±–∫–∞ –∞–±–æ –≤—ñ–¥—Å—É—Ç–Ω—è –∫–æ—Ä–µ–ª—è—Ü—ñ—è –º—ñ–∂ –æ–±—Å—è–≥–æ–º –ø—Ä–æ–¥–∞–∂—ñ–≤ —ñ —Å–µ—Ä–µ–¥–Ω—ñ–º –ø—Ä–∏–±—É—Ç–∫–æ–º ‚Äî –≤–∏—Å–æ–∫–∞ –≤–∏—Ä—É—á–∫–∞ –Ω–µ –∑–∞–≤–∂–¥–∏ –æ–∑–Ω–∞—á–∞—î –≤–∏—Å–æ–∫–∏–π –ø—Ä–∏–±—É—Ç–æ–∫.")
```

```python
# –ì—Ä–∞—Ñ—ñ–∫ —Å–µ—Ä–µ–¥–Ω—å–æ–≥–æ –ø—Ä–∏–±—É—Ç–∫—É
plot_time_series(average_profit_per_day, '–°–µ—Ä–µ–¥–Ω—ñ–π –ø—Ä–∏–±—É—Ç–æ–∫ –∑–∞ –¥–µ–Ω—å', '–°–µ—Ä–µ–¥–Ω—ñ–π –ø—Ä–∏–±—É—Ç–æ–∫')
statistical(average_profit_per_day, '–°–µ—Ä–µ–¥–Ω—ñ–π –ø—Ä–∏–±—É—Ç–æ–∫')

# –ì—Ä–∞—Ñ—ñ–∫ —Å–µ—Ä–µ–¥–Ω—å–æ–≥–æ –ø—Ä–∏–±—É—Ç–∫—É –∑ –≤–∏–ª—É—á–µ–Ω–Ω—è–º –∞–Ω–æ–º–∞–ª—ñ–π
average_profit_per_day_no_anomaly = average_profit_per_day[average_profit_per_day < 300]
average_profit_per_day_no_anomaly = average_profit_per_day_no_anomaly[average_profit_per_day_no_anomaly > -250]
plot_time_series(average_profit_per_day_no_anomaly, '–°–µ—Ä–µ–¥–Ω—ñ–π –ø—Ä–∏–±—É—Ç–æ–∫ –∑–∞ –¥–µ–Ω—å (–±–µ–∑ –∞–Ω–æ–º–∞–ª—ñ–π)', '–°–µ—Ä–µ–¥–Ω—ñ–π –ø—Ä–∏–±—É—Ç–æ–∫')
statistical(average_profit_per_day_no_anomaly, '–°–µ—Ä–µ–¥–Ω—ñ–π –ø—Ä–∏–±—É—Ç–æ–∫ (–±–µ–∑ –∞–Ω–æ–º–∞–ª—ñ–π)')

# –û–±—á–∏—Å–ª–µ–Ω–Ω—è —Ç—Ä–µ–Ω–¥—É (–ú–ù–ö)
x = np.arange(len(average_profit_per_day_no_anomaly))
y = average_profit_per_day_no_anomaly.values
coeffs = np.polyfit(x, y, 1)  # –ü–µ—Ä—à–∏–π –∫–æ–µ—Ñ—ñ—Ü—ñ—î–Ω—Ç ‚Äî –Ω–∞—Ö–∏–ª
slope = coeffs[0]

# –í–∏—Å–Ω–æ–≤–æ–∫
print(f"\n–ù–∞—Ö–∏–ª –ª—ñ–Ω—ñ—ó —Ç—Ä–µ–Ω–¥—É = {slope:.4f}). –¶–µ –±–ª–∏–∑—å–∫–æ –¥–æ 0, –æ—Ç–∂–µ –∑–∞–≥–∞–ª–æ–º –∑–∞ —Ä–æ–∫–∞–º–∏ —Å–µ—Ä–µ–¥–Ω—ñ–π –ø—Ä–∏–±—É—Ç–æ–∫ –Ω–∞ –¥–µ–Ω—å —î —Å—Ç–∞–ª–∏–º.")
```
